# TFMBS MNIST Demo
This notebook demonstrates end-to-end ternary quantization and Fabric offload for a simple MNIST model.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from pytfmbs.torch_integration import TFMBSLinear, TFMBSSequential

# 1. Define a simple model
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
)

print("Model defined.")
```

## 2. Offload to TFMBS
We can use `torch.compile` with the `tfmbs` backend for transparent offloading.

```python
from pytfmbs import torch_backend
torch_backend.register()

# Compile the model
compiled_model = torch.compile(model, backend="tfmbs")

print("Model compiled with TFMBS backend.")
```
