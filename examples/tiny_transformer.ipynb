# Tiny Transformer TFMBS Offload
Accelerating Attention blocks using Ternary Fabric.

```python
import torch
import torch.nn as nn
from pytfmbs.torch_integration import TFMBSLinear

class TinyAttention(nn.Module):
    def __init__(self, dim=64):
        super().__init__()
        self.qkv = nn.Linear(dim, dim * 3)
        self.proj = nn.Linear(dim, dim)
        
    def forward(self, x):
        # In real usage, this would be replaced or compiled
        return self.proj(self.qkv(x))

model = TinyAttention()
print("Tiny Transformer block ready.")
```
